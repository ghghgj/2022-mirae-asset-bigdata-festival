```python
import pandas as pd
df= pd.read_csv('/content/drive/Shareddrives/미래에셋/산업재해/산업재해.csv')
```

```python
new=[]
for i in df['사업장명\n( 현장명 )']:
    new.append(i.split('\n')[0])
```

```python
df['사업장명\n( 현장명 )']=new
```

```python
df.rename(columns = {"사업장명\n( 현장명 )": "사업장명"}, inplace = True)
df.rename(columns = {"재해율\n(%)": "재해율(%)"}, inplace = True)
df.rename(columns = {"규모별\n동종\n업종\n평균\n재해율\n(%)": "규모별 동종 업종 평균 재해율"}, inplace = True)
```

```python
df
```

```python
df['사업장명']= df['사업장명'].str.replace('㈜','')
df['사업장명']= df['사업장명'].str.replace('( 주 )','')
df['사업장명']= df['사업장명'].str.replace('( 원청 )','')
df['사업장명']= df['사업장명'].str.replace('( 하청 )','')
df['사업장명']= df['사업장명'].str.replace('()','')
df['사업장명']= df['사업장명'].str.replace(')','')
df['사업장명']= df['사업장명'].str.replace('(','')
df['사업장명']= df['사업장명'].str.replace('주식회사','')
df['사업장명']= df['사업장명'].str.replace('유 ','')
df['사업장명']= df['사업장명'].str.replace('유한회사','')
```

```python
# row 생략 없이 출력
pd.set_option('display.max_rows', None)
```

```python
rate= df[['사업장명','재해율(%)', '규모별 동종 업종 평균 재해율']]
```

```python
rate= rate.rename(columns={'사업장명':'CO_NM'})
rate
```

```python
rate['CO_NM']=rate['CO_NM'].str.lstrip()
rate['CO_NM']=rate['CO_NM'].str.rstrip()
```

```python
rate['CO_NM']
```

```python
df2 = pd.DataFrame(rate['CO_NM'].str.split(' ').tolist())

rate['CO_NM']=df2[0]
```

```python
rate.to_csv('/content/drive/Shareddrives/미래에셋/산업재해/재해율.csv')
```

```python
train= pd.read_csv("/content/drive/Shareddrives/미래에셋/data/train_업종.csv")
test =pd.read_csv("/content/drive/Shareddrives/미래에셋/data/test_업종.csv")
```

```python
from pandas.io.parsers.readers import read_csv
construction_train= train[train['업종명']==8]
construction_test= test[test['업종명']==8]
rate= read_csv('/content/drive/Shareddrives/미래에셋/산업재해/재해율.csv')
```

```python
rate
```

```python
con_train= pd.merge(rate, construction_train, on='CO_NM')
con_test= pd.merge(rate, construction_test, on='CO_NM')
```

```python
con_train=con_train.drop_duplicates(['CO_NM'], keep = 'first')
con_train=con_train[['CO_NM','CODE','재해율(%)','규모별 동종 업종 평균 재해율']]
con_train.to_csv('/content/drive/Shareddrives/미래에셋/병주/train_재해율.csv')

con_test=con_test.drop_duplicates(['CO_NM'], keep = 'first')
con_test=con_test[['CO_NM','CODE','재해율(%)','규모별 동종 업종 평균 재해율']]
con_test.to_csv('/content/drive/Shareddrives/미래에셋/병주/test_재해율.csv')
```
